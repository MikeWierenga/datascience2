### Graph isomorphism
Graphs may look different but they can be the same in the data structure. this notion of similarity is formalized through what is known as graph isomorphism

### definitions, theorem
- __definition2.7:__ consider two graphs G = (V,E) and G* = (V*, E*). G and G* are __isomorphic__ if there exists a one-to-one mapping : V -> V* such that every edge  E with e = (u,v), there is a unique edge e*  E* with e* = ((u), (v))
- __theorem 2.3:__ if two graphs G and G* are isomorphic, then their respective ordered degree sequence should be the same

## Graph Laplacian
- a matrix that describes the structure of a graph
- looks at how the nodes are connected to eachother and how information or signals can flow through the network
- the laplacian matrix is derived from the adjacency matrix
    - L(laplacian matrix) = D(diagonal matrix) - A(adjacency matrix)
        - D is __diagonal matrix 
        

## spectral graph theory
- uses the eignevalues and eigenvectors of adjacency matrix to study properties of the graph
- __eigenvectors__ can be used to find clusters but there are some issues
    - eigenvectors are sensitive to changes in the graph structure, which can make them less robust for clustering than other methods
    - hard to determine the number of clusters
    - results can depend on the initial conditions used for the algorithm
    - computationally expensive, and may not scale well on large graphs

## eigenvalues and eigenvectors of laplacian
- smallest eigenvalue of L is 0
- if there are n connected components in the graph, then there are n eigenvalues with value 0 
- first eigenvector column is constant
    - sequence of eigenvalues indicates a communitie
