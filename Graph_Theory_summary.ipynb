<h2>Jargon</h2>

<li><b>Walk</b>: A sequence of connected nodes
<li><b>Trail</b>: A walk with unique edges  
<li><b>Path</b>: A Trail with unique nodes  
<li><b>Closed walk</b>: A sequence of connected nodes but with the same start and end   
<li><b>Tour</b>: A walk with unique edges but with the same start and end  
<li><b>cycle</b>: A Trail with unique nodes but with the same start and end  <br><br>
    
<li><b>Eulerian</b>: An Eulerian graph traverses every edge in a graph exactly once, but may repeat edges
<li><b>Hamilton</b>: An hamilton graph visits each node in a graph exactly once but may repeat edges
<li><b>Complete graph</b>: A graph where every node is connected via an edge

<p><b>Di graphs</b>:<br>

<p><b>Reachability</b>:<br>
    <li>Reachability means that there exist a path from node u to node v</li>
    <li><b>Strongly connected</b> every node is reachable from all other nodes</li>
    <li><b>Weakly connected</b>: Not every noe is reachable from all other nodes</li>
</p>

<p><b>Toplogical sorting</b>:<br>
    <li>Toplogical sorting is an ordering of the nodes in a directed graph where for each directed edge from node a to node b, node a appears before node b in the ordering</li>
</p>
